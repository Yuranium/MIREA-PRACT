2.3.1 Задание
Найдите евклидово расстояние между двумя Series (точками) a и b, не
используя встроенную формулу.


```python
from math import sqrt
import pandas as pd
a = pd.Series([2, 4, 6, 8, 10])
b = pd.Series([20, 40, 60, 60, 100])
def evk_S(y, z):
   res = 0
   for i in range(len(y)):
       res += (y[i] - z[i]) ** 2
   return round(sqrt(res),2)
print("Евклидово расстояние: ",evk_S(a, b))
```

3.3.2 Задание
Загрузить фрейм данных по ссылке:
https://raw.githubusercontent.com/akmand/datasets/master/iris.csv.
Необходимо выполнить нормализацию первого числового признака
(sepal_length_cm) с использованием минимаксного преобразования, а
второго (sepal_width_cm) с задействованием z-масштабирования.


```python
import numpy as np
import pandas as pd
from scipy.stats import zscore
from sklearn import preprocessing
a = 'https://raw.githubusercontent.com/akmand/datasets/master/iris.csv'
dataframe = pd.read_csv(a) #читаем фрейм данных
x = np.array(dataframe['sepal_length_cm']) #делаем первый признак в массив
n = preprocessing.normalize([x]) #нормализация по методу минимакс
print("Нормализация по методу минимакс: ")
print(n,"\n","\n")
y = np.array(dataframe['sepal_width_cm']) #делаем второй признак в массив
m = zscore(y) #нормализация по методу Z-масштабирования
print("Нормализация по методу Z-масштабирования: ")
print(m)
```
