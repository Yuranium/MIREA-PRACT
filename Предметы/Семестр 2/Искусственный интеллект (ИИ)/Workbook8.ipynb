```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

url = r'https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv'

x = pd.read_csv(url)

plt.scatter(x.iloc[:,1], x.iloc[:,2])
plt.xlabel("1 кластер")
plt.show()


kmeans = KMeans(n_clusters=5, random_state=0)
clucters = kmeans.fit_predict(x.iloc[:,:4])
kmeans.cluster_centers_.shape
plt.scatter(x.iloc[:,1],x.iloc[:,2], s=40, c=kmeans.labels_ , cmap='tab10')
plt.xlabel("5 кластеров")
plt.show()



kmeans = KMeans(n_clusters=10, random_state=0)
clucters = kmeans.fit_predict(x.iloc[:,:4])
kmeans.cluster_centers_.shape
plt.scatter(x.iloc[:,1],x.iloc[:,2], s=40, c=kmeans.labels_ , cmap='tab10')
plt.xlabel("10 кластеров")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import AgglomerativeClustering



# загрузим набор данных
url = r'https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv'
customer_data = pd.read_csv(url)
print(customer_data.head(5),'\n')


data = customer_data.iloc[:, 2:4].values
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(28,12), dpi=180)
plt.figure(figsize=(10,7))
plt.title('ДЕНДОГРАММА на основе 2 признаков')
dend = shc.dendrogram(shc.linkage(data, method='ward'))


# группировка точек данных в 4 кластера
cluster = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
print("Группировка точек данных в 4 кластера:",'\n','\n', cluster.fit_predict(data))


plt.figure(figsize=(10,7))
plt.title('Как на самом деле были сгруппированы наши данные (всего 4 признака)')
plt.scatter(data[:,0], data[:,1], c=cluster.labels_,s=50, cmap='rainbow')
```
