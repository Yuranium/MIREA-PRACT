```python
import math as m

class Calculator_trigonometria():
    def cos(self, degree):
        return m.cos(degree)

    def sin(self, degree):
        return m.sin(degree)

    def tan(self, degree):
        return m.tan(degree)

    def asin(self, degree):
        return m.asin(degree)

    def acos(self, degree):
        return m.acos(degree)

    def atan(self, degree):
        return m.atan(degree)

    def toRadians(self, degree):
        return m.radians(degree)


clc = Calculator_trigonometria()
a = int(input(" : "))
print(" - " , clc.sin(a))
print(" - " , clc.cos(a))
print(" - ", clc.tan(a))
print(" - ", clc.asin(a))
print(" - " , clc.acos(a))
print(" - " , clc.atan(a))
print("  ", clc.toRadians(a))
```

```python
class Tree:
    def __init__(self, kids, next=None):
        self.kids = self.val = kids
        self.next = next
        
t = Tree(Tree('a', Tree('b', Tree('d', 'e'))), Tree('c', Tree('f')))

print(' -  ', t.kids.val)

print('  - ', end='')
print(t.kids.next.val, end=', ')
print(t.kids.next.next.val, end=', ')
print(t.kids.next.next.next)

print('  -  ', end='')
print(t.next.kids, end=', ')
print(t.next.next.kids)
```

```python
class Tree:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data
    
    def Print_Tree(self):
        if self.left:
            self.left.Print_Tree()
        print(self.data)
        if self.right:
            self.right.Print_Tree()
    
    def Vstavka(self, item):
        if self.data:
            if item < self.data:
                if self.left is None:
                    self.left = Tree(item)
                else:
                    self.left.Vstavka(item)
            elif item > self.data:
                if self.right is None:
                    self.right = Tree(item)
                else:
                    self.right.Vstavka(item)
        else:
            self.data = item

            
tree = Tree(7)

tree.Vstavka(1)
tree.Vstavka(2)

tree.Vstavka(8)
tree.Vstavka(9)

tree.Print_Tree()
```

```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split


x = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
target = [0, 0, 0, 1, 1, 1]

dataset = pd.DataFrame(x)
y = pd.DataFrame(target)


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)


classifier = tree.DecisionTreeClassifier()
classifier.fit(x_train, y_train)

#   
tree.plot_tree(classifier)
# 
y_pred = classifier.predict(x_test)


print(" :\n\n",confusion_matrix(y_test, y_pred),"\n")
print(" (  ):\n\n",classification_report(y_test, y_pred, zero_division=0))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics

url = r'https://raw.githubusercontent.com/likarajo/petrol_consumption/master/data/petrol_consumption.csv'
dataset = pd.read_csv(url)
dataset.describe().round(3)




#   
print(dataset.shape)
dataset.describe()

#   
plt.scatter(dataset['Petrol_tax'], dataset['Average_income'], color='b', label='Average income')
plt.xlabel(' ')
plt.ylabel('  ')
plt.show()


x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
#print(x)
#print(y)
print(' ')

# ,      , ,         .
#       ,   -   
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

#      
#    LinearRegression,       fit()
regressor = DecisionTreeRegressor()
regressor.fit(x_train, y_train)


tree.plot_tree(regressor)

#  
y_pred = regressor.predict(x_test)
y_pred

#          :
df = pd.DataFrame({"": y_test, "": y_pred})
print("        : ")
print(df.head())
print(' ')

#       :
print("  MSE: ", metrics.mean_squared_error(y_test, y_pred))
print("   MAE: ", metrics.mean_absolute_error(y_test, y_pred))
print("%   MAE: ", (metrics.mean_absolute_error(y_test, y_pred) / np.average(y) * 100).round(3))

# MSE ,        .
# MAE        .
#  MAE ,       
```
