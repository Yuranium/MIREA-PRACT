Задайте 4 точки в трехмерном пространстве, рассчитайте между ними расстояния по описанным в примере выше метрикам. Отобразите точки в трехмерном пространстве.

```python
import matplotlib.pyplot as plt
import numpy as np

f = plt.figure() #создаем поле для графика
a = f.add_subplot(111, projection='3d') #создаем трехмерный график
a.scatter(0, 0, 0) #создаем первую точку
A = np.array([0, 0, 0]) #создаем массив из первой точки
a.scatter(1, 1, 1) #создаем вторую точку
B = np.array([1, 1, 1]) #создаем массив из второй точки
a.scatter(2, 2, 2) #создаем третью точку
C = np.array([2, 2, 2]) #создаем массив из третий точки
a.scatter(3, 3, 3) #создаем четвертую точку
D = np.array([3, 3, 3]) #создаем массив из четвертой точки
plt.xlabel("4 точки в трехмерном пространстве")
plt.show()
print("Расстояние Евклида точек A и B:",round(np.linalg.norm(A-B),2))
print("Квадрат расстояния Евклида точек A и B:",round(np.linalg.norm(A-B) ** 2,2))
print("Расстояние Чебышева между точками A и B:",round(np.linalg.norm(A-B, ord=np.inf),2))
print("Расстояния Хемминга между точками A и B:",round(np.linalg.norm(A-B, ord=1),2))
print("Расстояние Евклида точек A и D:",round(np.linalg.norm(A-D),2))
print("Квадрат расстояния Евклида точек A и D:",round(np.linalg.norm(A-D) ** 2,2))
print("Расстояние Чебышева между точками A и D:",round(np.linalg.norm(A-D, ord=np.inf),2))
print("Расстояния Хемминга между точками A и D:",round(np.linalg.norm(A-D, ord=1),2))
print("Расстояние Евклида точек A и C:",round(np.linalg.norm(A-C),2))
print("Квадрат расстояния Евклида точек A и C:",round(np.linalg.norm(A-C) ** 2,2))
print("Расстояние Чебышева между точками A и C:",round(np.linalg.norm(A-C, ord=np.inf),2))
print("Расстояние Хемминга между точками A и D:",round(np.linalg.norm(A-C, ord=1)),2)
```

2.3.1 Задание
Для предыдущего примера поэкспериментируйте с параметрами
классификатора:
1. Установите другое количество ближайших соседей (k = 1, 5, 10).
2. Установите размер тестовой выборки 15% от всего датасета.
3. Постройте графики и оцените качество моделей, проанализируйте
результаты.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
from sklearn.metrics import accuracy_score
iris = sns.load_dataset('iris')
print(iris)

x_train, x_test, y_train, y_test = train_test_split(
    iris.iloc[:, :-1], # берём все колонки кроме последней в признаки
    iris.iloc[:, -1],   # последнюю в целевую перменную (класс)
    test_size=0.15      #размер тестовой выборки 15%
)
# обучим метод трех ближайших соседей
model = KNeighborsClassifier(n_neighbors=10)
model.fit(x_train, y_train)

# получим предсказания модели
y_pred = model.predict(x_test)

#покажем на графике, что отражает полученное число
#красным цветом обозначены точки, для которых классификация сработала неправильно
plt.figure(figsize=(10, 7))

sns.scatterplot(
    data=iris,
    x='petal_width', y='petal_length',
    hue='species',
    s=70
)
plt.xlabel('Длина лепестка, см')
plt.ylabel('Ширина лепестка, см')
plt.legend(loc=2)
plt.grid()


#перебираем все объекты из теста
for i in range(len(y_test)):
    # если предсказание неправильное
    if np.array(y_test)[i] != y_pred[i]:
        # то подсвечиваем точку красным
        plt.scatter(x_test.iloc[i, 3], x_test.iloc[i, 2], color='red', s=150)

print(f'Точность: {accuracy_score(y_test, y_pred) :.3}')
```

3.3.2 Задание
Определите набор признаков человека, по аналогии из РТ 1, – например,
цвет глаз и конвертируйте его в матрицу признаков.

```python
from sklearn.feature_extraction import DictVectorizer

person_dict = [{"tall": 2, "Blue_eye": 1, "Black": 2}, #создаем словарь
             {"tall": 5, "asian": 4, "Black": 3},
             {"Blue_eye": 1, "Black": 3, "asian": 1}]
dict_vectorizer = DictVectorizer(sparse=False) #создаем вектор словаря
priznaki = dict_vectorizer.fit_transform(person_dict)   #конвертируем в матрицу признаков
print("Матрица признаков:")
print(priznaki)
```
